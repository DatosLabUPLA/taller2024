```python
from bs4 import BeautifulSoup as sp
import requests
import pandas as pd
from io import StringIO
```

```python
url = "http://imantofagasta.cl/transparencia/paginas/Remunera.aspx?id=2"
```

```python
req = requests.get(url)
```

```python
soup = sp(req.text, 'html.parser')
```

```python
tabla = soup.find("table", id="ctl00_ContentPlaceHolder1_gvContrata")

data = []

for row in tabla.find_all('tr'):
    row_data = []
    for cell in row.find_all(['th', 'td']):
        row_data.append(cell.get_text())
    data.append(row_data)
```

```python
df = pd.DataFrame(data[1:], columns=data[0])
```

```python
df.head(3)
```

```python
for col in df.columns:
    df[col] = df[col].str.replace('\n', '', regex=True)
```

```python
df.to_csv("Municipalidad_Antofagasta.csv")
```
