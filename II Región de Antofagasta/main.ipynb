```python
from bs4 import BeautifulSoup as sp
import requests
import pandas as pd
from io import StringIO
```

```python
url1 = "http://imantofagasta.cl/transparencia/paginas/Remunera.aspx?id=2"
url2 = "https://calamatransparencia.insico.cl/paginas/Remunera.aspx?id=2"
```

```python
def convertScrapperToDf(url, idtable = "ctl00_ContentPlaceHolder1_gvContrata"):
    req = requests.get(url)
    soup = sp(req.text, 'html.parser')
    tabla = soup.find("table", id=idtable)
    data = []
    
    for row in tabla.find_all('tr'):
        row_data = []
        for cell in row.find_all(['th', 'td']):
            row_data.append(cell.get_text().strip())
        data.append(row_data)
    df = pd.DataFrame(data[1:], columns=data[0])
    return df
```

```python
df_antof = convertScrapperToDf(url1)
```

```python
df_antof.to_csv("Municipalidad_Antofagasta.csv")
```
